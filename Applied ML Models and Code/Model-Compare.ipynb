```python
import matplotlib.pyplot as plt
models = ["LR", "RFC", "SVC", "XGBClassifier", "DecisionTreeClassifier", 
          "GradientBoostingClassifier", "KNeighborsClassifier", "AdaBoostClassifier", 
          "SGDClassifier", "LinearSVC", "RoBERTa-GCN"]

def bar_positions(base_index, bar_width, n_metrics, group_width, n_models):
    return [base_index * group_width + i * bar_width for i in range(n_metrics)]
bar_width = 0.1
group_width = bar_width * 8 + 0.2

opacity = 0.8

accuracy = [0.91, 0.86, 0.95, 0.90, 0.96, 0.89, 0.94, 0.86, 0.79, 0.91, 0.98]
precision = [0.90, 0.87, 0.94, 0.91, 0.95, 0.89, 0.94, 0.85, 0.80, 0.89, 0.98]
recall = [0.91, 0.86, 0.95, 0.87, 0.92, 0.91, 0.92, 0.89, 0.79, 0.91, 0.97]
mcc = [0.45, 0.50, 0.65, 0.55, 0.45, 0.47, 0.62, 0.45, 0.48, 0.35, 0.93]
sensitivity = [0.91, 0.85, 0.94, 0.90, 0.93, 0.88, 0.92, 0.87, 0.81, 0.92, 0.99]
specificity = [0.33, 0.43, 0.49, 0.64, 0.33, 0.23, 0.30, 0.31, 0.45, 0.35, 0.25]
ppv = [0.91, 0.89, 0.91, 0.89, 0.93, 0.90, 0.91, 0.86, 0.78, 0.92, 0.97]
npv = [0.78, 0.86, 0.65, 0.78, 0.79, 0.80, 0.84, 0.85, 0.80, 0.84, 0.98]


fig, ax = plt.subplots(figsize=(20, 10))

for i in range(len(models)):
    positions = bar_positions(i, bar_width, 8, group_width, len(models))
    ax.bar(positions[0], accuracy[i], bar_width, alpha=opacity, color='b', label='Accuracy' if i == 0 else "")
    ax.bar(positions[1], precision[i], bar_width, alpha=opacity, color='g', label='Precision' if i == 0 else "")
    ax.bar(positions[2], recall[i], bar_width, alpha=opacity, color='r', label='Recall' if i == 0 else "")
    ax.bar(positions[3], mcc[i], bar_width, alpha=opacity, color='y', label='MCC' if i == 0 else "")
    ax.bar(positions[4], sensitivity[i], bar_width, alpha=opacity, color='c', label='Sensitivity (SEN)' if i == 0 else "")
    ax.bar(positions[5], specificity[i], bar_width, alpha=opacity, color='m', label='Specificity (SPE)' if i == 0 else "")
    ax.bar(positions[6], ppv[i], bar_width, alpha=opacity, color='orange', label='Positive Predictive Value (PPV)' if i == 0 else "")
    ax.bar(positions[7], npv[i], bar_width, alpha=opacity, color='purple', label='Negative Predictive Value (NPV)' if i == 0 else "")


ax.set_xlabel('Models')
ax.set_ylabel('Scores')
ax.set_title('Comprehensive Model Performance Comparison with RoBERTa-GCN')
ax.set_xticks([i * group_width + 3.5 * bar_width for i in range(len(models))])
ax.set_xticklabels(models, rotation=45, ha="right")
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), shadow=True)
plt.subplots_adjust(right=0.8)

fig.savefig('comprehensive_model_performance_comparison_with_legend_right.png', format='png')
plt.xticks(fontsize='large')  
plt.show()
```
